# Seaborn
Seaborn is a Python data visualization library based on Matplotlib that provides a highâ€‘level interface for drawing attractive and informative statistical graphics.

```python
# Mock Data
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

data = {
    "Person": [f"Person {c}" for c in "ABCDEFGHIJ"],
    "Age": np.random.randint(20, 50, size=10),
    "Steps": np.random.randint(4000, 15000, size=10),
    "Calories": np.random.randint(1500, 3500, size=10)
}

df = pd.DataFrame(data)
print(df)
```

## Visualization
This part vislualizes the data frame

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.barplot(data=df,  x="Person", y="Steps")
plt.title("Daily Steps per Person")
plt.xlabel("Person")
plt.ylabel("Steps")
plt.show()
```
