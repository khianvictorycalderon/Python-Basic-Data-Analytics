# Pandas
Pandas is a fast, powerful, flexible, and easy-to-use open-source data analysis and manipulation tool, built on top of the Python programming language.

```python
import pandas as pd

# Create a simple Series
data = [10, 20, 30, 40]
s = pd.Series(data)

print(s)
```

```python
import numpy as np

data = np.array([
    ["month", "income", "spending"],
    ["January", 4200, 3500],
    ["February", 4300, 3600],
    ["March", 4100, 3200],
    ["April", 4500, 3800],
    ["May", 4700, 4000],
    ["June", 4600, 3900],
    ["July", 4400, 3700],
    ["August", 4800, 4100],
    ["September", 4500, 3800],
    ["October", 4300, 3600],
    ["November", 4600, 3900],
    ["December", 5000, 4200]
])
```

**Converting to DataFrame (skip header row)**

```python
# Data Frame
data_content = data[1:] # Starting from the first index
columns = data[0] # Just the first index
df = pd.DataFrame(data_content, columns=columns)

# Converting to numerical values first
df["income"] = pd.to_numeric(df["income"])
df["spending"] = pd.to_numeric(df["spending"])

print(df)
```

*Data Manipulation*

```python
# Adding a savings column
df["savings"] = df["income"] - df["spending"]
income_tax_rate = 0.18 # 18%, fictional only
df["income_after_tax"] = df["income"] * (1 - income_tax_rate)

# Shows the data frame with new columns
print(df)
```

## Some Aggregated functions:
- `sum()`
- `count()`
- `min()`
- `max()`
- `median()`
- `std()`
- `var()`
- `mode()`

```python
## Averages
print(f"Average Income: {df["income"].mean()}")
print(f"Average Spending: {df["spending"].mean()}")
print(f"Average Savings: {df["savings"].mean()}")
print(f"Average Income After Tax: {df["income_after_tax"].mean()}")
```

## Visualization
This part visualizes the current data frame

```python
# Visualization of all columns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
months = range(0, 12)
months_labels = [
    "Jan", "Feb", "Mar", "Apr",
    "May", "Jun", "Jul", "Aug",
    "Sep", "Oct", "Nov", "Dec"
]

plt.plot(months, df["income"], marker="o", label="Income", color="green")
plt.plot(months, df["spending"], marker="o", label="Spending", color="red")
plt.plot(months, df["savings"], marker="o", label="Savings", color="blue")
plt.plot(months, df["income_after_tax"], marker="o", label="Income after Tax", color="orange")

plt.xlabel("Months")
plt.ylabel("Amount ($)")
plt.xticks(months, months_labels)
plt.legend()
plt.grid(True)

plt.show()
```
