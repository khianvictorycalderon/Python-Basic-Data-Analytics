## Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
It provides an object-oriented API for embedding plots into applications and a MATLAB-like interface via pyplot for quick plotting.

### Matplotlib visualizations

```python
# Sample Mock Data to visualize
import numpy as np

# 12 months of the year
months_numbers = np.arange(1, 13) # 1 to 12
months_labels = [
    "Jan", "Feb", "March", "Apr",
    "May", "Jun", "Jul", "Aug",
    "Sep", "Oct", "Nov", "Dec"
]

# Average income per household per month (in dollars)
income = np.array([
    4200, 4300, 4150, 4400,
    4500, 4600, 4550, 4700,
    4800, 5000, 5100, 5250
])

# Average spending per household per month (in dollars)
spending = np.array([
    3100, 3200, 3150, 3300,
    3400, 3450, 3500, 3550,
    3650, 3750, 3800, 3950
])
```

```python
# Plotting their income
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(months_numbers, income, marker="o", label="Income", color="green")

plt.title("Average Income of household per Month")
plt.xlabel("Month")
plt.xticks(months_numbers, months_labels) # For each label, replaces (1-12) with (Jan - Dec)
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting their spending

plt.figure(figsize=(10, 5))
plt.plot(months_numbers, spending, marker="o", label="Spending", color="red")

plt.title("Average Spending of household per Month")
plt.xlabel("Month")
plt.xticks(months_numbers, months_labels)
plt.ylabel("Amount ($)")
plt.legend()
plt.grid(True)
plt.show()
```
